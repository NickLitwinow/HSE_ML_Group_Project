```python
from datetime import datetime

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.simplefilter(action='ignore', category=(FutureWarning, UserWarning))

pd.set_option("display.max_columns", 100)
```

```python
df = pd.read_csv("../train_final.csv")
```

```python
# Настройки графиков
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# --- 1. is_repeated_guest ---
plt.figure()
sns.countplot(data=df, x='is_repeated_guest')
plt.title("Распределение: Постоянный гость или нет")
plt.xlabel("is_repeated_guest")
plt.ylabel("Количество")
plt.show()
```

```python
# --- 2. previous_cancellations ---
plt.figure()
sns.histplot(data=df, x='previous_cancellations', bins=20, kde=True)
plt.title("Распределение: Количество предыдущих отмен")
plt.xlabel("previous_cancellations")
plt.ylabel("Частота")
plt.show()
```

```python
# --- 3. previous_bookings_not_canceled ---
plt.figure()
sns.histplot(data=df, x='previous_bookings_not_canceled', bins=20, kde=True)
plt.title("Распределение: Предыдущие успешные брони")
plt.xlabel("previous_bookings_not_canceled")
plt.ylabel("Частота")
plt.show()
```

```python
# --- 4. reserved vs assigned room type ---
plt.figure()
sns.countplot(data=df, x='reserved_room_type', order=sorted(df['reserved_room_type'].unique()))
plt.title("Типы забронированных номеров")
plt.xlabel("reserved_room_type")
plt.ylabel("Количество")
plt.show()
```

```python
plt.figure()
sns.countplot(data=df, x='assigned_room_type', order=sorted(df['assigned_room_type'].unique()))
plt.title("Типы назначенных номеров")
plt.xlabel("assigned_room_type")
plt.ylabel("Количество")
plt.show()
```

```python
# --- 5. Совпадение номеров ---
df['room_match'] = df['reserved_room_type'] == df['assigned_room_type']

plt.figure()
sns.countplot(data=df, x='room_match')
plt.title("Совпадение забронированного и назначенного типа номера")
plt.xlabel("Совпадает ли номер")
plt.ylabel("Количество")
plt.show()
```
